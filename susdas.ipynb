## PSO

```python
def func(k):
  return math.sin(k+3) - math.sin(k+1) - (2 * math.cos(k+2))

# Define PSO algo
def pso_algo(n, m, c1, c2, v, x, r1, r2, omega):
  fitness = [func(i) for i in x]
  pbest = x
  gbest = x[fitness.index(min(fitness))]
  pbest_fitness = fitness
  gbest_fitness = func(gbest)

  for i in range(m):
    fitness = [func(i) for i in x]
    for j in range(n):
      if fitness[j] < pbest_fitness[j]:
        pbest[j] = x[j]
        pbest_fitness[j] = fitness[j]
      if fitness[j] < gbest_fitness:
        gbest = x[j]
        gbest_fitness = fitness[j]
  for k in range(n):
    v[k] = (omega * v[k]) + (c1 * r1[i] * (pbest[k] - x[k])) + (c2 *
            r2[i] * (gbest - x[k]))
    x[k] += v[k]
  return gbest
```

```python
num_iter = 10
num_par = 100
omega = 0.5
r1 = [random.random() for i in range(num_par)]
r2 = [random.random() for i in range(num_par)]
c1 = 1
c2 = 1
v = [0.0 for i in range(num_par)]
x = [random.randint(-100,100) for i in range(num_par)]
```

```python
res = pso_algo(num_par, num_iter, c1, c2, v, x, r1, r2, omega)
print("Nilai x minimum : {}".format(res))
print("Nilai f(x) : {}".format(func(res)))
```
